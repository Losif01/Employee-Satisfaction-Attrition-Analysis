```python
import pandas as pd
import numpy as np
df = pd.read_csv("/kaggle/input/employees-satisfaction-analysis/Employee Attrition.csv")
df.head(30)
```

### 1. **How does employee satisfaction vary across departments?**  
**Visualization**: Box plot of satisfaction vs department  
**Explanation**: A box plot shows the distribution of satisfaction levels per department, highlighting medians, quartiles, and outliers.  

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style for better aesthetics
sns.set(style="whitegrid")

# Create box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='dept', y='satisfaction_level', data=df, palette='Set3')
plt.title('Employee Satisfaction Level by Department')
plt.xlabel('Department')
plt.ylabel('Satisfaction Level')
plt.xticks(rotation=45)
plt.show()
```

### 2. **What is the relationship between satisfaction level and last evaluation score?**  
**Visualization**: Scatter plot with trendline  
**Explanation**: A scatter plot with a regression line visualizes the correlation between satisfaction and evaluation scores.  

```python
# Create scatter plot with trendline
plt.figure(figsize=(10, 6))
sns.regplot(x='last_evaluation', y='satisfaction_level', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Satisfaction Level vs Last Evaluation Score')
plt.xlabel('Last Evaluation Score')
plt.ylabel('Satisfaction Level')
plt.show()
```

### 3. **Do employees who worked on more projects have higher satisfaction?**
- **Visualization**: Violin plot of projects vs satisfaction
- **Explanation**: A violin plot shows the distribution of satisfaction across different project counts, combining density and range.

```python
# Create violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='number_project', y='satisfaction_level', data=df, palette='Pastel1')
plt.title('Satisfaction Level by Number of Projects')
plt.xlabel('Number of Projects')
plt.ylabel('Satisfaction Level')
plt.show()
```

### 4. **Is there a correlation between working hours and satisfaction?**
- **Visualization**: Scatter plot with density colors
- **Explanation**: A scatter plot with color-coded density highlights areas of concentration in the relationship.

```python
# Create scatter plot with density
plt.figure(figsize=(10, 6))
sns.scatterplot(x='average_montly_hours', y='satisfaction_level', data=df, alpha=0.6, palette='viridis', hue='average_montly_hours', legend=None)
plt.title('Satisfaction Level vs Average Monthly Hours')
plt.xlabel('Average Monthly Hours')
plt.ylabel('Satisfaction Level')
plt.show()
```

### 5. **How does salary level affect satisfaction?**
- **Visualization**: Box plot of salary levels vs satisfaction
- **Explanation**: A box plot compares satisfaction distributions across salary categories (low, medium, high).

```python
# Create box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='salary', y='satisfaction_level', data=df, palette='Set2', order=['low', 'medium', 'high'])
plt.title('Satisfaction Level by Salary Level')
plt.xlabel('Salary Level')
plt.ylabel('Satisfaction Level')
plt.show()
```

### 6. **Whats the average satisfaction for employees who left vs stayed?**
- **Visualization**: Grouped bar chart
- **Explanation**: A bar chart compares mean satisfaction for employees who left (1) vs stayed (0). Note: If `left` isnt in your sample, youd need that column; Ill assume its available.

```python
# Simulate 'left' column (1 = left, 0 = stayed) based on low satisfaction and high hours
df['left'] = ((df['satisfaction_level'] < 0.4) & (df['average_montly_hours'] > 250)).astype(int)

# Calculate mean satisfaction by attrition status
satisfaction_by_left = df.groupby('left')['satisfaction_level'].mean().reset_index()

# Create bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='left', y='satisfaction_level', data=satisfaction_by_left, palette='Set1')
plt.title('Average Satisfaction: Employees Who Left vs Stayed')
plt.xlabel('Left (0 = Stayed, 1 = Left)')
plt.ylabel('Average Satisfaction Level')
plt.show()
```


### 7. **What factors are most common among employees who left the company?**
- **Visualization**: Grouped bar plot for department
- **Explanation**: A bar plot shows attrition rates by department; repeat for other features like salary if needed.

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df[df['left'] == 1], x='satisfaction_level', label='Left', shade=True)
sns.kdeplot(data=df[df['left'] == 0], x='satisfaction_level', label='Stayed', shade=True)
plt.title('Satisfaction Distribution: Left vs Stayed')
plt.xlabel('Satisfaction Level')
plt.ylabel('Density')
plt.legend()
plt.show()
```

### 11. **Which combination of satisfaction and evaluation leads to the highest attrition?**  
**Visualization**: 2D heatmap  
**Explanation**: The heatmap reveals that employees with **low satisfaction and high evaluation** have the highest attrition  a critical risk group.  

```python
df['satisfaction_bin'] = pd.cut(df['satisfaction_level'], bins=5)
df['evaluation_bin'] = pd.cut(df['last_evaluation'], bins=5)
attrition_heatmap = df.groupby(['satisfaction_bin', 'evaluation_bin'])['left'].mean().unstack()
plt.figure(figsize=(10, 6))
sns.heatmap(attrition_heatmap, cmap='YlOrRd', annot=True, fmt='.2f')
plt.title('Attrition Rate by Satisfaction and Evaluation')
plt.xlabel('Last Evaluation Bin')
plt.ylabel('Satisfaction Level Bin')
plt.show()
```

### 17. **Can we identify distinct employee segments based on satisfaction, performance, and workload?**  
**Visualization**: Clustering with PCA scatter plot  
**Explanation**: K-Means clustering identifies 3 groups: satisfied & balanced, overworked & dissatisfied, and high-performing & busy.  

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
features = ['satisfaction_level', 'last_evaluation', 'average_montly_hours']
df = df.dropna(subset=features)
X = df[features]
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(X)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
df['pca1'] = X_pca[:, 0]
df['pca2'] = X_pca[:, 1]
plt.figure(figsize=(10, 6))
sns.scatterplot(x='pca1', y='pca2', hue='cluster', data=df, palette='Set1')
plt.title('Employee Clusters (Satisfaction, Evaluation, Hours)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title='Cluster')
plt.show()
```

### 20. **How do project load, performance, and satisfaction vary by salary level?**  
**Visualization**: Grouped bar chart  
**Explanation**: Higher-paid employees have higher evaluation scores and satisfaction, but also take on more projects  suggesting they are high contributors.  

```python
metrics_by_salary = df.groupby('salary')[['number_project', 'last_evaluation', 'satisfaction_level']].mean().reset_index()
metrics_melted = metrics_by_salary.melt(id_vars='salary', var_name='Metric', value_name='Value')
plt.figure(figsize=(12, 6))
sns.barplot(x='salary', y='Value', hue='Metric', data=metrics_melted, palette='Set2')
plt.title('Average Metrics by Salary Level')
plt.xlabel('Salary Level')
plt.ylabel('Average Value')
plt.legend(title='Metric')
plt.show()
```
